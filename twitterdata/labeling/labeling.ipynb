## 필터링 된 데이터셋 합치기

```python
대마초 = pd.read_csv("대마초.csv")
대마초_필터링 = pd.read_csv("대마초_필터링.csv")
인디카 = pd.read_csv("인디카.csv")
케타민 = pd.read_csv("케타민.csv")
차가운술 = pd.read_csv("차가운술.csv")
작대기 = pd.read_csv("작대기.csv")
```
